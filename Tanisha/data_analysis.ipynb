# Squeezing Measurement
*Squeezing Level (dB)*  
$= -10 \log_{10} ( \frac{\Delta X^2}{\Delta X^2_{\text{vacuum}}})$  

- $\Delta X= \text{squeezed state noise variance}$  
- $\Delta X{vaccum}= \text{vaccum state noise variance}$



# Polarization Measurements

By the menthod described by Peter in elog 3147, the Polarization measurements were made for the beam incident on the crystal from the optical fibre. The following were found- 
- ~81 % linear polarization for the 1064 nm beam
- ~93 % linear polarization for the 532 nm beam 

By rotating the half waveplate, the beam was made incident on the optical fibre and at all angles there was non-zero reflected and refracted intensity, showing elliptical polarization. This is due to instrinsic losses in the fibre that may be due to- 
- Imperfect alignment or residual mode mixing at the ending strand of the fibre that couples it to the crystal
- Bending of the fibre causing mechanical strain


# Non-Linear Gain Measuremt of the crystal 
First the 1064nm beam was used to align the optics and ensure that the beam is coupled well with the PPLN crystal via the optical fiber. The 1064 nm beam on passing through the crystal produces a 532nm beam via SHG. Higher the   intensity of beam generated after SHG, higher will be the squeezing.  
On taking the SHG measurents under the conditions of- 0° angle of incidence on the PPLN crystal face, optimal temperature of 52.5°C (repoted in elog 3147), the non-linear power gain was found to be-   

$P_\text{in}=9.5 mW$ (1064 nm)   
$P_\text{out}=123 {\mu} W$ (SHG)   
$\text{Power Gain}$= $-10 \cdot \log_{10} \left( \frac{P_{\text{out}}}{P_{\text{in}}} \right)$ =$-41.12 dB$

 ### Gain Balancing- Substraction of Shot noise from PDs
 To suppress common-mode noise (e.g.- Residual noise from laser, electronics,shot noise from PDs) the following steps are performed-
 - With ~0.6mW of 1064nm LO power hitting each PD, and no 1064nm power from the signal (or squeezing) path, the AC BNC for each PD goes into an input for the Moku:Lab.
 - Then the values in the PD matrixes in the Moku's FIR filter builder module, are tweeked till on subtracting the two detector signals (North PD & South PD) → flat trace close to zero is obtained.
- The plot in attachment 1 is from the gain balanced stage (Residual trace ~0)-  
    * North PD's matrix value is set to **1.2**.
    * South PD's matrix value is set to **-0.9**. (ref- elog 3164)

As, evident in the plot- The flat trace could be obtained in the higher frequency domain of $3\text{MHz}-7\text{MHz}$. The subtraction is then bandpassed between $3\text{MHz}-7\text{MHz}$ and sent to Moku:Go (which is physically a BNC cable connecting from the Moku:Lab out1 into Moku:GO In1) after applying suffient gain to ensure that the moku:go receives a large enough signal. On substracting the shot noises electronically, some subsidual noise remains in the gain balanced state mostly due to innate noise in MOKU.
The residual RMS noise is found to be: ≈ $1.2 e-6 \frac{v}{\sqrt{Hz}}$

```python
# Filter between 3 MHz and 7 MHz
frequency1 = combined_df['% Frequency (Hz)'].values
frequency2=combined_df_n['% Frequency (Hz)'].values
mask1 = (frequency1 >= 3e6) & (frequency1 <= 7e6)
mask2 = (frequency2 >= 3e6) & (frequency2 <= 7e6)

filtered_freq = frequency1[mask1]
filtered_exp = exp[mask2]
filtered_combined = combined_df[mask1]

# Calculate shot noise level (mean in flat region)
noise_level = filtered_combined[' Input 1 (V/rtHz)'].mean()

# Plot
plt.figure(figsize=(10, 6))
plt.plot(filtered_freq, filtered_combined[' Input 1 (V/rtHz)'].values, label=' experimental gain balanced state', color='green')
plt.plot(filtered_freq, filtered_exp, label='Subtracted Trace', color='black')
plt.axhline(y=noise_level, color='purple', linestyle='--', label=f'Noise Level ≈ {noise_level:.2e} V/√Hz')

# Labels and title
plt.xlabel('Frequency (Hz)')
plt.ylabel('Input (V/√Hz)')
plt.title('Noise Level in 3–7 MHz Band')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
